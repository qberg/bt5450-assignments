```python
import matplotlib.pyplot as plt # for using plots
import seaborn as sns # for using 2D maps
import pandas as pd # for using csv file (data analysis)
import numpy as np
```

```python
df = pd.read_csv('Cancergeometry.csv') # to open file
df.head() 
```

```python
print(df.columns)
```

```python
dfm=df[df['diagnosis']=='M']
dfb=df[df['diagnosis']=='B']
```

```python
# boxplot for geometric significance

y=df.diagnosis
list = ['id','diagnosis'] #assigned for excluding
features = df.drop(list,axis = 1,inplace = False) #Feature assigned inplace false indicates to exclude
data_plot= pd.concat([y,features.iloc[:,:]],axis=1)
data_plot=pd.melt(data_plot,id_vars="diagnosis",var_name="features",value_name='value')
plt.figure(figsize=(10,30))
sns.boxplot(x="value",y="features",hue="diagnosis",data=data_plot)
plt.title("Box plot for geometrical significance")
plt.xscale('log')

#violin plot for geometric significance
plt.figure(figsize=(10,30))
ax = sns.violinplot(x="value", y="features", hue="diagnosis",data=data_plot, palette="muted")
plt.title("Violin plot for geometrical significance")
plt.xscale('log')
```

**5a. The box and violin plots for comparing the geometrical significance has been plotted.**  

```python
#Comparing the geometric aspects of tumour using scatter plot
#using scatter matrix

data_scatter_matrix= pd.concat([y,features.iloc[:,:10]],axis=1)
pd.plotting.scatter_matrix(data_scatter_matrix, alpha=0.5,diagonal='kde',figsize=(30, 30))
plt.show()
```

```python
# error bar plots to visualize the mean and sd of 
# perimeter mean, concave point mean, texture se and smoothness worst stratified by tumour type

dfm=df[df['diagnosis']=='M']
dfb=df[df['diagnosis']=='B']
variables= ['perimeter_mean','concave points_mean', 'texture_se', 'smoothness_worst']
# perimeter mean
M_perimeter_mean= np.mean(dfm['perimeter_mean'])
B_perimeter_mean= np.mean(dfb['perimeter_mean'])
names= ['Malignant', 'Benign']
perimeter_means= [M_perimeter_mean,B_perimeter_mean]

#Standard error= SD/sqrt(sample size)

Stderrorpm_M= np.std(dfm['perimeter_mean'])/np.sqrt(len(dfm.index))
Stderrorpm_B= np.std(dfb['perimeter_mean'])/np.sqrt(len(dfb.index))
error= [Stderrorpm_M, Stderrorpm_B]
plt.bar(names, perimeter_means)
plt.errorbar(names,perimeter_means , yerr=error, fmt= 'o', color= 'r')
plt.title("Error Bar plot for perimeter mean")
plt.show()

#concave point mean
M_concave_mean= np.mean(dfm['concave points_mean'])
B_concave_mean= np.mean(dfb['concave points_mean'])
names= ['Malignant', 'Benign']
concave_points_means= [M_concave_mean,B_concave_mean]

Stderrorcpm_M= np.std(dfm['concave points_mean'])/np.sqrt(len(dfm.index))
Stderrorcpm_B= np.std(dfb['concave points_mean'])/np.sqrt(len(dfb.index))
error= [Stderrorcpm_M, Stderrorcpm_B]
plt.bar(names, concave_points_means)
plt.errorbar(names,concave_points_means, yerr=error, fmt= 'o', color= 'r')
plt.title("Error Bar plot for concave point mean")
plt.show()

#Texture se
M_texture_se= np.mean(dfm['texture_se'])
B_texture_se= np.mean(dfb['texture_se'])
names= ['Malignant', 'Benign']
texture_ses= [M_texture_se,B_texture_se]

Stderrorts_M= np.std(dfm['texture_se'])/np.sqrt(len(dfm.index))
Stderrorts_B= np.std(dfb['texture_se'])/np.sqrt(len(dfb.index))
error= [Stderrorts_M, Stderrorts_B]
plt.bar(names, texture_ses)
plt.errorbar(names,texture_ses, yerr=error, fmt= 'o', color= 'r')
plt.title("Error Bar plot for texture standard error")
plt.show()

#Smoothness worst
M_smoothness_worst= np.mean(dfm['smoothness_worst'])
B_smoothness_worst= np.mean(dfb['smoothness_worst'])
names= ['Malignant', 'Benign']
smoothness_worsts= [M_concave_mean,B_concave_mean]

Stderrorsw_M= np.std(dfm['smoothness_worst'])/np.sqrt(len(dfm.index))
Stderrorsw_B= np.std(dfb['smoothness_worst'])/np.sqrt(len(dfb.index))
error= [Stderrorsw_M, Stderrorsw_B]
plt.bar(names, smoothness_worsts)
plt.errorbar(names,smoothness_worsts, yerr=error, fmt= 'o', color= 'r')
plt.title("Error Bar plot for smoothness worst")
plt.show()
```

**5d. Observation on the error bar plot**

From the above error plots, in which the malignant and benign types are stratified it is clear that the mean values have less spread in the benign tumor, and the variance is high for the malignant type.

From the texture SE, it indicates the standard error for both benign and malignant tumours are nearly similar.
The smoothness worst indicates the smoothness of the malignant tumour have more variance with extreme values, and that of benign is less variant when compared to the malignant types.

```python
# histogram and empirical cumulative distribution
# for  â€˜smoothness meanâ€™, â€˜compactness meanâ€™ and â€˜fractal dimension worstâ€™

variables= ['smoothness_mean', 'compactness_mean', 'fractal_dimension_worst']

x_sm=dfm['smoothness_mean']
y_sm= dfb['smoothness_mean']
bins = np.linspace(0, 0.20, 20)
plt.figure(figsize=(8,5))
plt.hist([x_sm,y_sm], bins, alpha=0.5, label=['Malignant', 'Benign'])
plt.ylabel("Frequency")
plt.xlabel("Smoothness_mean")
plt.title("Histogram for the values of smoothness mean")
plt.grid(True)
plt.legend(loc='upper right')
plt.show()

#Compactness mean
x_cm=dfm['compactness_mean']
y_cm= dfb['compactness_mean']
bins = np.linspace(0, 0.20, 20)
plt.figure(figsize=(8,5))
plt.hist([x_cm,y_cm], bins, alpha=0.5, label=['Malignant', 'Benign'])
plt.ylabel("Frequency")
plt.xlabel("Compactness_mean")
plt.title("Histogram for the values of Compactness mean")
plt.grid(True)
plt.legend(loc='upper right')
plt.show()

#Fractal_dimension_worst
x_fdw=dfm['fractal_dimension_worst']
y_fdw= dfb['fractal_dimension_worst']
bins = np.linspace(0, 0.20, 20)
plt.figure(figsize=(8,5))
plt.hist([x_fdw,y_fdw], bins, alpha=0.5, label=['Malignant', 'Benign'])
plt.ylabel("Frequency")
plt.xlabel("Fractal_dimension_worst")
plt.title("Histogram for the values of Fractal dimension worst")
plt.grid(True)
plt.legend(loc='upper right')
plt.show()
```

```python
# empirical cumulative distribution
# for  â€˜smoothness meanâ€™, â€˜compactness meanâ€™ and â€˜fractal dimension worstâ€™
#variables= ['smoothness_mean', 'compactness_mean', 'fractal_dimension_worst']

def ECDF(data):
    x_val= np.sort(data)
    size=x_val.size
    y_val= np.arange(1,size+1) / size
    return (x_val,y_val)

#smoothness mean
xm_sm,ym_sm= ECDF(dfm['smoothness_mean'])
xb_sm, yb_sm= ECDF(dfb['smoothness_mean'])

fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(xm_sm[:], ym_sm[:], s=10, c='r', marker="x", label='Malignant')
ax1.scatter(xb_sm[:],yb_sm[:], s=10, c='b', marker="o", label='Benign')
plt.legend(loc='upper left');
plt.title("ECDF plot to show the proportion of mean smoothness")
plt.xlabel("Smoothness mean")
plt.ylabel("Proportion")
plt.show()

#compactness mean
xm_cm,ym_cm= ECDF(dfm['compactness_mean'])
xb_cm, yb_cm= ECDF(dfb['compactness_mean'])

fig = plt.figure()
ax2 = fig.add_subplot(111)

ax2.scatter(xm_cm[:], ym_cm[:], s=10, c='r', marker="x", label='Malignant')
ax2.scatter(xb_cm[:],yb_cm[:], s=10, c='b', marker="o", label='Benign')
plt.legend(loc='upper left');
plt.title("ECDF plot to show the proportion of mean compactness")
plt.xlabel("Compactness mean")
plt.ylabel("Proportion")
plt.show()

#fractal dimension worst
xm_fdw,ym_fdw= ECDF(dfm['fractal_dimension_worst'])
xb_fdw, yb_fdw= ECDF(dfb['fractal_dimension_worst'])

fig = plt.figure()
ax3 = fig.add_subplot(111)

ax3.scatter(xm_fdw[:], ym_fdw[:], s=10, c='r', marker="x", label='Malignant')
ax3.scatter(xb_fdw[:],yb_fdw[:], s=10, c='b', marker="o", label='Benign')
plt.legend(loc='upper left');
plt.title("ECDF plot to show the proportion of Worst fractal dimension")
plt.xlabel("Fractal dimesnion worst")
plt.ylabel("Proportion")
plt.show()
    
```

**5e. Difference in the information provided by the relative histograms and ECDF plots.**

*A histogram provides the value with high frequency, whereas the ECDF can show the outliers value and the range where most value points lies in.*

i. Smoothness mean
In the histogram, since a normal distribution of the values are seen. The benign type of tumor has high frequency at the mean value between the range of 0.08-0.09 and the malignant type with the mean value at the rang eof 0.09 to 0.11. 
In the ECDF, the two plots are almost parallel with malignant type having higher smoothness mean value than benign type

ii. Compactness mean
The histogram of the malignant type is skewed to the right (Negative skew) indicating that more tumors of this type are having high compactness mean, whereas benign tumor is slightly left skewed (positive skew), making it an assymmetric plot.
The ECDF shows that the compactness mean of malignant type is mostly less than 0.25 and is less than 0.15 for benign type

iii. Fractal dimension worst
Worst values are symmetrical in the histogram, and the values are less than 0.16 for both malignant and tumor. But that does mean, there is a high spread/ variance in the values.
